```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os

# Base URL
base_url = "https://www.moneycontrol.com/financials/adanienterprises/balance-sheetVI/AE13#AE13"

# Number of pages you want to scrape (in this case, 5)
num_pages = 2

# Generate URLs
urls = [f"{base_url}{i}#AE13" for i in range(1, num_pages+1)]

# Function to scrape data from a single URL
def scrape_table(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    table = soup.find('table', {'class': 'mctable1'})  # Adjust this selector if needed

    years = []
    current_assets = []
    current_liabilities = []

    for row in table.find_all('tr'):
        cols = row.find_all('td')
        if cols:
            row_title = cols[0].text.strip().lower()

            if "total current liabilities" in row_title:
                years = [header.text.strip() for header in table.find_all('th')[1:]]  # Extract year headers
                current_liabilities = [col.text.strip() for col in cols[1:]]
            elif "total current assets" in row_title:
                current_assets = [col.text.strip() for col in cols[1:]]
    print(years,row_title)
    return years, current_assets, current_liabilities

# Create an empty list to store the results
results = []

# Scrape data for each URL
for i, url in enumerate(urls, 1):
    print(f"Scraping data for Page {i}...")

    years, total_current_assets, total_current_liabilities = scrape_table(url)

    for year, assets, liabilities in zip(years, total_current_assets, total_current_liabilities):
        results.append({
            'Page': i,
            'Year': year,
            'Total Current Assets': assets,
            'Total Current Liabilities': liabilities
        })

    print(f"Data for Page {i} collected.")

# Convert results to a DataFrame and save as a CSV
df_results = pd.DataFrame(results)
# df_results.to_csv('bsheet.csv', index=False)

# print("Summary data has been scraped and saved.")
df_results
```
